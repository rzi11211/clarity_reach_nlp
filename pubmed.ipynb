```python
from Bio import Entrez
```

```python
def search(query):
    Entrez.email = 'storytellinginsights@gmail.com'
    handle = Entrez.esearch(db='pubmed',
                            sort='relevance',
                            retmax='20',
                            retmode='xml',
                            term=query)
    results = Entrez.read(handle)
    return results
```

```python
def fetch_details(id_list):
    ids = ','.join(id_list)
    Entrez.email = 'storytellinginsights@gmail.com'
    handle = Entrez.efetch(db='pubmed',
                           retmode='xml',
                           id=ids)
    results = Entrez.read(handle)
    return results
```

```python
if __name__ == '__main__':
    results = search('machine learning' -datetype PDAT -mindate 2020 -maxdate 2021 -sort 'Relevance')
    id_list = results['IdList']
    papers = fetch_details(id_list)
    for i, paper in enumerate(papers['PubmedArticle']):
        print("{}) {}".format(i+1, paper['MedlineCitation']['DateCompleted']))
        print("{}) {}".format(i+1, paper['MedlineCitation']['Article']['ArticleTitle']))
        print("{}) {}".format(i+1, paper['MedlineCitation']['Article']['Abstract']['AbstractText']))
        
```
